# 6.) The company you are working for wants to target at a False Positive rate of 5% what threshold should you use? (Use oversampled data)

_Predicting fraud when it is not fraud._

```python
from sklearn.metrics import confusion_matrix
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import roc_curve
import matplotlib.pyplot as plt
```

```python
clf = LogisticRegression()
model = clf.fit(train_data_x, train_data_y)

probs = model.predict_proba(test_data_x)

threshold = 0
for t in np.arange(0, 1, 0.01):
  preds = probs[:,1] > t
  cm = confusion_matrix(test_data_y, preds)
  fpr = cm[0,1] / (cm[0,0] + cm[0,1])
  if fpr <= 0.05:
    threshold = t
    break
```

```python
print(f"The threshold is {threshold}.")
```

# 7.) If the company makes .02 * amt on True transactions and loses -amt on False (Use original data)

```python
test_data_x
```

```python
test_data_x["Preds"] = preds
```

```python
test_data_x["signal"] = test_data_x["Preds"] * test_data_y
```

```python
df_compare = pd.concat([test_data_x, test_data_y], axis = 1)
```

```python
df_compare["Preds"] = [0 if i == False else 1 for i in df_compare["Preds"]]
```

```python
df_compare["FN"] = (df_compare["is_fraud"] == 1) & (df_compare["Preds"] == 0)
```

```python
df_compare
```

```python
df_compare["profit"] = df_compare["signal"] * 0.02 * df_compare["amt"] + df_compare["FN"] * -df_compare["amt"]
```
